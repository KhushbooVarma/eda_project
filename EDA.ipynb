```python
columns_to_check = ['price', 'sqft_living', 'bedrooms', 'bathrooms']

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for relevant columns
Q1 = df_imp[columns_to_check].quantile(0.25)
Q3 = df_imp[columns_to_check].quantile(0.75)

print("Q1 (25th percentile):")
print(Q1)
print("\nQ3 (75th percentile):")
print(Q3)

```

```python
IQR = Q3 - Q1

print("\nInterquartile Range (IQR):")
print(IQR)
```

```python
# Identify outliers
outliers = ((df_imp[['price', 'sqft_living', 'bedrooms', 'bathrooms']] < (Q1 - 1.5 * IQR)) | 
            (df_imp[['price', 'sqft_living', 'bedrooms', 'bathrooms']] > (Q3 + 1.5 * IQR))).any(axis=1)

print("\nOutliers identified:")
print(df_imp[outliers])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_imp is already defined as a DataFrame

# Define bins for sqft_living
bins = [0, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000]
labels = ['<1000', '1000-1500', '1500-2000', '2000-2500', '2500-3000', '3000-3500', 
          '3500-4000', '4000-4500', '4500-5000']

# Categorize sqft_living into bins
df_imp['sqft_living_range'] = pd.cut(df_imp['sqft_living'], bins=bins, labels=labels, right=False)

# Calculate mode price for each sqft_living range
mode_price = df_imp.groupby('sqft_living_range')['price'].apply(lambda x: x.mode()[0]).reset_index()

# Plotting the bar chart for mode price
plt.figure(figsize=(10, 6))
sns.barplot(x='sqft_living_range', y='price', data=mode_price, palette='viridis')
plt.title('Mode Price by Square Foot Living Area')
plt.xlabel('Square Foot Living Area')
plt.ylabel('Mode Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
